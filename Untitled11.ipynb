```python
import pandas as pd

# Verinizi yükleyin
data = pd.read_csv(file_path)  # 'file_path.csv'yi dosyanızın yolu ile değiştirin

# 'lunch' sütunundaki eşsiz değerleri bulun
unique_values = data['lunch'].unique()

# Benzersiz değerleri yazdırın
print("Benzersiz değerler:", unique_values)
```

```python
# 'lunch' sütunundaki her türdeki kişi sayısını bulun
lunch_counts = data['lunch'].value_counts()

# Sonuçları yazdırın
print("Lunch türlerine göre kişi sayıları:")
print(lunch_counts)
```

```python
average_scores = data.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Sonuçları yazdırın
print("Cinsiyete göre ortalama skorlar:")
print(average_scores)
```

```python
# 'race/ethnicity' sütununa göre gruplama ve her grup için ortalamaları hesaplama
average_scores_by_race = data.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Sonuçları yazdırın
print("Race/Ethnicity'ye göre ortalama skorlar:")
print(average_scores_by_race)
```

```python
# 'parental level of education' sütununa göre gruplama ve her grup için ortalamaları hesaplama
average_scores_by_education = data.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()

# Sonuçları yazdırın
print("Parental level of education'a göre ortalama skorlar:")
print(average_scores_by_education)
```

```python
# 'lunch' sütununa göre gruplama ve her grup için ortalamaları hesaplama
average_scores_by_lunch = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()

# Sonuçları yazdırın
print("Lunch türlerine göre ortalama skorlar:")
print(average_scores_by_lunch)
```
